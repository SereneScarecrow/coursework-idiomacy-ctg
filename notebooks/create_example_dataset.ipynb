```python
import pandas as pd

with open("C:/Users/user/coursework-idiomacy-ctg/data/example/example.txt", 'r', encoding='utf-8') as f:
    example = f.read().split('***')
    example = [i.strip().split('\n') for i in example]
    example_dict = {i[0].lower(): i[1:] for i in example}
    
example_dict
```

```python
example_data = pd.DataFrame(example_dict)
example_data
```

```python
example_data.to_csv("C:/Users/user/coursework-idiomacy-ctg/data/example/example.csv")
```
